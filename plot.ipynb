```python
def PlotRes():
    ImageIter = iter(data_iter)
    count = 0
    for X, Y in ImageIter:
        fig = px.imshow(np.array(Y[0]),zmax=140, color_continuous_scale="gray_r", width=600, height=600, title=" PM10 ")
        x = [11, 29, 47, 65, 83]
        y = [14, 32, 50, 68, 86]
        xticks=["115E -", "120E -", "125E -", "130E -", "135E -"]
        yticks=["50N -", "45N -", "40N -", "35N -", "30N -"]  
        fig.update_yaxes(tickvals=y, ticktext=yticks)
        fig.update_xaxes(tickvals=x, ticktext=xticks, tickangle=-90)
        fig.write_image("Result/True" + str(count) + ".png", scale=2)

        pred, _ = RNN(X.to(device))
        fig = px.imshow(pred.cpu().detach().numpy()[0],zmax=140, color_continuous_scale="gray_r", width=600, height=600, title=" PM10 ")
        x = [11, 29, 47, 65, 83]
        y = [14, 32, 50, 68, 86]
        xticks=["115E -", "120E -", "125E -", "130E -", "135E -"]
        yticks=["50N -", "45N -", "40N -", "35N -", "30N -"]  
        fig.update_yaxes(tickvals=y, ticktext=yticks)
        fig.update_xaxes(tickvals=x, ticktext=xticks, tickangle=-90)
        fig.write_image("Result/Pred" + str(count) + ".png", scale=2)

        count += 1
```
